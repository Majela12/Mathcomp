```
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

# Datos
x = np.array([0, 1, 2, 3, 4, 5, 6]).reshape(-1, 1)
y = np.array([-0.9, 0, 2, 4.5, 8.3, 13, 18])

# Regresión Polinómica (grado 2)
poly_features = PolynomialFeatures(degree=2)
X_poly = poly_features.fit_transform(x)
model_poly = LinearRegression().fit(X_poly, y)
y_pred_poly = model_poly.predict(X_poly)

# Coeficiente de correlación para regresión polinómica
r_value_poly = np.corrcoef(y.flatten(), y_pred_poly)[0, 1]

# Regresión Lineal Múltiple (usando las mismas características)
# Aquí usamos solo x como predictor, para lineal múltiple agregaríamos más variables
model_multi = LinearRegression().fit(x, y)
y_pred_multi = model_multi.predict(x)

# Coeficiente de correlación para regresión lineal múltiple
r_value_multi = np.corrcoef(y.flatten(), y_pred_multi)[0, 1]

# Resultados
print("Regresión Polinómica:")
print(f"Coeficientes: {model_poly.coef_}, Intercepto: {model_poly.intercept_}")
print(f"Coeficiente de correlación (r): {r_value_poly}\n")

print("Regresión Lineal Múltiple:")
print(f"Coeficiente: {model_multi.coef_}, Intercepto: {model_multi.intercept_}")
print(f"Coeficiente de correlación (r): {r_value_multi}")

# Graficar resultados
plt.figure(figsize=(12, 6))

# Graficar Regresión Polinómica
plt.subplot(1, 2, 1)
plt.scatter(x, y, label='Datos Reales', color='red')
plt.plot(x, y_pred_poly, color='blue', label='Ajuste Polinómico (Grado 2)')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Ajuste Polinómico de Segundo Grado')
plt.legend()
plt.grid()

# Graficar Regresión Lineal Múltiple
plt.subplot(1, 2, 2)
plt.scatter(x, y, label='Datos Reales', color='red')
plt.plot(x, y_pred_multi, color='green', label='Ajuste Lineal Múltiple')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Ajuste Lineal Múltiple')
plt.legend()
plt.grid()

plt.tight_layout()
plt.show()
```
