```python
import itertools
import math
from math import comb
from math import factorial
from itertools import permutations
```

```python
total_libros = factorial(3) // factorial(3 - 2)
print(f"Ejercicio 14: {total_libros}")
```

```python
total_productos = comb(8, 4)
print(f"Ejercicio 16: {total_productos}")
```

```python
buenos = comb(20, 4)
defectuosos = comb(5, 2)
total_interruptores = buenos * defectuosos
print(f"Ejercicio 17: {total_interruptores}")
```

```python
total_canicas = factorial(9) // (factorial(3) * factorial(2) * factorial(4))
print(f"Ejercicio 18: {total_canicas}")
```

```python

titulos = list(range(1, 11))  # Representamos los títulos como números del 1 al 10
permutaciones = list(permutations(titulos, 3))
print(f"Ejercicio 19: {len(permutaciones)}")
```

```python

a = factorial(9) == 9 * 8 * factorial(6)
b = factorial(5) * factorial(4) == factorial(20)
c = factorial(5) + factorial(5) == factorial(10)
d = factorial(8) == factorial(9) // 9
e = factorial(5) == factorial(10) // factorial(2)
f = (3/4 + 1/4) == factorial(1)
g = factorial(3) * factorial(2) == factorial(6)
h = factorial(8) == factorial(6) * 56

print(f"Ejercicio 20a: {a}")
print(f"Ejercicio 20b: {b}")
print(f"Ejercicio 20c: {c}")
print(f"Ejercicio 20d: {d}")
print(f"Ejercicio 20e: {e}")
print(f"Ejercicio 20f: {f}")
print(f"Ejercicio 20g: {g}")
print(f"Ejercicio 20h: {h}")
```
