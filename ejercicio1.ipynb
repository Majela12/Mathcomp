```
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

# Datos
x1 = np.array([1, 1, 2, 3, 1, 2, 3, 3]).reshape(-1, 1)
x2 = np.array([-1, 0, 0, 1, 1, 2, 2, 0]).reshape(-1, 1)
y = np.array([1.6, 3, 1.1, 1.2, 3.2, 3.3, 1.7, 0.1])

# Regresión Lineal Múltiple
X_multi = np.hstack((x1, x2))  # Combinar x1 y x2
model_multi = LinearRegression().fit(X_multi, y)
y_pred_multi = model_multi.predict(X_multi)
r_value_multi = np.corrcoef(y, y_pred_multi)[0, 1]

# Resultados Regresión Lineal Múltiple
print("Regresión Lineal Múltiple:")
print(f"Coeficientes: {model_multi.coef_}, Intercepto: {model_multi.intercept_}")
print(f"Coeficiente de correlación (r): {r_value_multi}")

# Regresión Polinómica (grado 2)
poly_features = PolynomialFeatures(degree=2)
X_poly = poly_features.fit_transform(X_multi)
model_poly = LinearRegression().fit(X_poly, y)
y_pred_poly = model_poly.predict(X_poly)
r_value_poly = np.corrcoef(y, y_pred_poly)[0, 1]

# Resultados Regresión Polinómica
print("\nRegresión Polinómica:")
print(f"Coeficientes: {model_poly.coef_}, Intercepto: {model_poly.intercept_}")
print(f"Coeficiente de correlación (r): {r_value_poly}")

# Graficar resultados
plt.figure(figsize=(14, 6))

# Graficar Regresión Lineal Múltiple
plt.subplot(1, 2, 1)
plt.scatter(range(len(y)), y, label='Datos Reales', color='red')
plt.plot(range(len(y)), y_pred_multi, color='blue', label='Predicción Múltiple')
plt.xlabel('Índice')
plt.ylabel('Y')
plt.title('Regresión Lineal Múltiple')
plt.legend()

# Graficar Regresión Polinómica
plt.subplot(1, 2, 2)
plt.scatter(range(len(y)), y, label='Datos Reales', color='red')
plt.plot(range(len(y)), y_pred_poly, color='blue', label='Predicción Polinómica')
plt.xlabel('Índice')
plt.ylabel('Y')
plt.title('Regresión Polinómica')
plt.legend()

plt.tight_layout()
plt.show()
```
